```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe 
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
##create an instance of Onehotencoder
encoder=OneHotEncoder()
```

```python
## perform fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

encoded: এটি হল fit_transform() ফাংশনের মাধ্যমে পাওয়া এক্সট্র্যাক্ট করা ফলাফল, যা একাধিক শ্রেণীর জন্য One-Hot Encoding প্রয়োগ করে (যেমন color কলামে red, blue, এবং green থাকলে, প্রতিটি শ্রেণীকে বাইনারি (১ অথবা ০) রূপে রূপান্তরিত করে)।

pd.DataFrame(encoded, columns=encoder.get_feature_names_out()):

pd.DataFrame(): এটি পাণ্ডাস ডেটাফ্রেম তৈরি করে।

encoded: এখানে এটি সেই এক্সট্র্যাক্ট করা ডেটা, যা আমরা OneHotEncoder ব্যবহার করে রূপান্তরিত করেছি।

columns=encoder.get_feature_names_out(): এই অংশটি কলামগুলোর নাম সেট করে। encoder.get_feature_names_out() ফাংশনটি OneHotEncoder দ্বারা রূপান্তরিত কেটাগরি গুলোর নাম প্রদান করে, যেমন: color_red, color_blue, color_green ইত্যাদি।

```python
## ORdinal Encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
# create a sample dataframe with an ordinal variable
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```
