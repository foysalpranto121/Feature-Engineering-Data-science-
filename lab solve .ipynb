```python
from collections import deque

graph = {
    "D": ["B"],
    "B": ["A", "E"],
    "A": ["C"],
    "E": ["F"],
    "F": ["C"],
    "C": []
}

def bfs_goal(start, goal):
    visited = set()
    queue = deque([start])
    order = []

    while queue:
        node = queue.popleft()

        if node not in visited:
            visited.add(node)
            order.append(node)

            
            if node == goal:
                return order

            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return order   

print("BFS  D to F:", bfs_goal("D", "F"))
```
